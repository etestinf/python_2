```python
import requests
import ipywidgets as widgets
from IPython.display import display, clear_output
from datetime import datetime

# Google Forms konfigurācija
form_url = "https://docs.google.com/forms/d/e/1FAIpQLSdTSrnEP9GQajJRRCKqQMbwWGS_TL3o3B4T_fequkWmUZQCCg/formResponse"
entry_ids = {
    "Students": "entry.1649720525",
    "Datums": "entry.667959859",
    "Rezultats": "entry.940605140",
    "1": "entry.213159259",
    "2": "entry.267050949",
    "3": "entry.418893037",
    "4": "entry.1447996441",
    "5": "entry.1914428828",
    "6": "entry.1728217231",
    "7": "entry.1875638404",
    "8": "entry.907839017"
}

# Piemēra dati (parasti šie būtu testā izveidoti)
questions_data = [
    ("Ko dara pygame.init()?", ["..."], 0),
    ("Kura metode tiek izmantota...?", ["..."], 0),
    ("Kā tiek atjaunināts logs...?", ["..."], 0),
    ("Ko pārbauda event.type...?", ["..."], 0),
    ("Kurš modulis tiek izmantots...?", ["..."], 0),
    ("Kurš tkinter logelements...?", ["..."], 0),
    ("Ko dara mainloop()...?", ["..."], 0),
    ("Kurš tkinter logelements ļauj...?", ["..."], 0)
]
results = [10, 5, 10, 10, 7.5, 10, 10, 7.5]
attempts = [1, 2, 1, 1, 2, 1, 1, 2]
name_input = widgets.Text(value='Anna', description='Vārds:')

def send_to_google_form(_):
    student_name = name_input.value.strip() or "Nezināms"
    total = sum(r if r else 0 for r in results)
    final_score = round(total)
    now = datetime.now().strftime("%Y-%m-%d %H:%M")

    form_data = {
        entry_ids["Students"]: student_name,
        entry_ids["Datums"]: now,
        entry_ids["Rezultats"]: str(final_score) + "%"
    }

    # Iekļauj jautājuma tekstu + mēģinājumus + punktus
    for i in range(len(questions_data)):
        question_text = questions_data[i][0]
        attempt_count = attempts[i]
        score = results[i] if results[i] is not None else 0
        response = f"{question_text} (Mēģinājumi: {attempt_count}) – {score} punkti"
        form_data[entry_ids[str(i+1)]] = response

    try:
        response = requests.post(form_url, data=form_data)
        if response.status_code == 200:
            display(widgets.HTML("<h4 style='color:green;'>Rezultāti veiksmīgi nosūtīti uz Google Formu!</h4>"))
        else:
            display(widgets.HTML(f"<h4 style='color:red;'>Kļūda sūtot datus (statuss: {response.status_code})</h4>"))
    except Exception as e:
        display(widgets.HTML(f"<h4 style='color:red;'>Kļūda sūtot datus: {e}</h4>"))

send_button = widgets.Button(description="📤 Sūtīt uz Google Formu", button_style='primary')
send_button.on_click(send_to_google_form)
display(name_input, send_button)
```
