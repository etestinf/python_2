**Instrukcijas studentiem:**

- Izlasi katru jautājumu uzmanīgi un izvēlies vienu pareizo atbildi.
- Par katru nepareizu mēģinājumu gala rezultāts samazinās par **5%**.
- Jautājumu secība un atbilžu kārtība ir nejauša katram mēģinājumam.
- Kad visi jautājumi ir atbildēti, nospied pogu **"Iesniegt testu"**, lai redzētu savu rezultātu.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import random

results = {}
attempts = {}

questions_data = [
    {
        "question": "1. Ko dara pygame.init()?",
        "options": ["Inicializē visus importētos pygame moduļus", "Palaiž spēles ciklu", "Ielādē spēles logu", "Izveido galveno varoņa sprite objektu"],
        "correct": "Inicializē visus importētos pygame moduļus"
    },
    {
        "question": "2. Kura metode tiek izmantota, lai uzzīmētu taisnstūri pygame logā?",
        "options": ["pygame.draw.rect()", "pygame.screen.rectangle()", "pygame.create.rect()", "pygame.render.rect()"],
        "correct": "pygame.draw.rect()"
    },
    {
        "question": "3. Kā tiek atjaunināts logs pygame pēc zīmēšanas?",
        "options": ["pygame.display.flip()", "pygame.update.display()", "pygame.screen.refresh()", "pygame.render.frame()"],
        "correct": "pygame.display.flip()"
    },
    {
        "question": "4. Ko pārbauda event.type == pygame.QUIT?",
        "options": ["Vai lietotājs ir nospiedis loga aizvēršanas pogu", "Vai lietotājs nospieda taustiņu Enter", "Vai spēlei jāsākas no jauna", "Vai spēlētājs zaudēja"],
        "correct": "Vai lietotājs ir nospiedis loga aizvēršanas pogu"
    },
    {
        "question": "5. Kurš modulis tiek izmantots, lai veidotu GUI lietotnes Python valodā?",
        "options": ["tkinter", "pygame.gui", "guimodule", "tkbuilder"],
        "correct": "tkinter"
    },
    {
        "question": "6. Kurš tkinter logelements tiek izmantots, lai saņemtu tekstu no lietotāja?",
        "options": ["Entry", "Label", "Button", "Canvas"],
        "correct": "Entry"
    },
    {
        "question": "7. Ko dara mainloop() tkinter lietotnē?",
        "options": ["Palaiž GUI notikumu ciklu", "Pieslēdzas datubāzei", "Atjauno Python vidi", "Izveido loga izvietojumu"],
        "correct": "Palaiž GUI notikumu ciklu"
    },
    {
        "question": "8. Kurš tkinter logelements ļauj zīmēt figūras kā taisnstūrus un tekstu?",
        "options": ["Canvas", "Frame", "Box", "Entry"],
        "correct": "Canvas"
    },
]

random.shuffle(questions_data)

question_widgets = []

def create_question_widget(qid, question, options, correct):
    out = widgets.Output()
    radio = widgets.RadioButtons(options=options.copy(), layout={'width': 'max-content'})
    button = widgets.Button(description='Pārbaudīt')
    feedback = widgets.HTML()
    random.shuffle(radio.options)

    def on_click(b):
        if qid not in attempts:
            attempts[qid] = 0
        attempts[qid] += 1
        if radio.value == correct:
            penalty = 0.05 * (attempts[qid] - 1)
            score = max(1 - penalty, 0)
            results[qid] = score
            feedback.value = "<span style='color:green'>✔️ Pareizi!</span>"
            radio.disabled = True
            button.disabled = True
        else:
            feedback.value = "<span style='color:red'>❌ Nepareizi. Mēģini vēlreiz.</span>"

    button.on_click(on_click)
    with out:
        display(widgets.HTML(f"<b>{question}</b>"), radio, button, feedback)
    return out

for i, q in enumerate(questions_data):
    q_widget = create_question_widget(i, q["question"], q["options"], q["correct"])
    question_widgets.append(q_widget)

submit_button = widgets.Button(description="✅ Iesniegt testu", button_style='success')
result_output = widgets.Output()

def calculate_score(_):
    with result_output:
        clear_output()
        if len(results) < len(questions_data):
            display(widgets.HTML("<b style='color:red'>⚠️ Lūdzu, atbildi uz visiem jautājumiem pirms iesniegšanas.</b>"))
            return
        final_score = round(sum(results.values()) / len(results) * 100)
        color = "green" if final_score >= 70 else "red"
        status = "✅ Nokārtots!" if final_score >= 70 else "❌ Nenokārtots."
        display(widgets.HTML(f"<h3>Rezultāts: <span style='color:{color}'>{final_score}%</span> – {status}</h3>"))

submit_button.on_click(calculate_score)

for widget in question_widgets:
    display(widget)
display(submit_button)
display(result_output)
```
