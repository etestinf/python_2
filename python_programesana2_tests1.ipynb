# 🧪 Python Kodēšana 2 – Tests (1.–4. nodarbība)

Izvēlies pareizās atbildes. Atkārtota mēģinājuma gadījumā rezultāts samazināsies par 5%.

**Pabeidzot**, spied uz "✅ Iesniegt testu", lai redzētu savu rezultātu.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import random

questions_data = [
    {
        "question": "1. Ko dara pygame.init()?",
        "options": ["Inicializē visus importētos pygame moduļus", "Palaiž spēles ciklu", "Ielādē spēles logu", "Izveido galveno varoņa sprite objektu"],
        "correct": 0
    },
    {
        "question": "2. Kura metode tiek izmantota, lai uzzīmētu taisnstūri pygame logā?",
        "options": ["pygame.draw.rect()", "pygame.screen.rectangle()", "pygame.create.rect()", "pygame.render.rect()"],
        "correct": 0
    },
    {
        "question": "3. Kā tiek atjaunināts logs pygame pēc zīmēšanas?",
        "options": ["pygame.display.flip()", "pygame.update.display()", "pygame.screen.refresh()", "pygame.render.frame()"],
        "correct": 0
    },
    {
        "question": "4. Ko pārbauda event.type == pygame.QUIT?",
        "options": ["Vai lietotājs ir nospiedis loga aizvēršanas pogu", "Vai lietotājs nospieda taustiņu Enter", "Vai spēlei jāsākas no jauna", "Vai spēlētājs zaudēja"],
        "correct": 0
    },
    {
        "question": "5. Kurš modulis tiek izmantots, lai veidotu GUI lietotnes Python valodā?",
        "options": ["tkinter", "pygame.gui", "guimodule", "tkbuilder"],
        "correct": 0
    },
    {
        "question": "6. Kurš tkinter logelements tiek izmantots, lai saņemtu tekstu no lietotāja?",
        "options": ["Entry", "Label", "Button", "Canvas"],
        "correct": 0
    },
    {
        "question": "7. Ko dara mainloop() tkinter lietotnē?",
        "options": ["Palaiž GUI notikumu ciklu", "Pieslēdzas datubāzei", "Atjauno Python vidi", "Izveido loga izvietojumu"],
        "correct": 0
    },
    {
        "question": "8. Kurš tkinter logelements ļauj zīmēt figūras kā taisnstūrus un tekstu?",
        "options": ["Canvas", "Frame", "Box", "Entry"],
        "correct": 0
    },
]

random.shuffle(questions_data)
results = [None] * len(questions_data)
attempts = [0] * len(questions_data)
outputs = []

def create_question_widget(index, qdata):
    out = widgets.Output()
    options = qdata['options'][:]
    correct_option = options[qdata['correct']]
    random.shuffle(options)
    correct_index = options.index(correct_option)

    radio = widgets.RadioButtons(options=options, layout={'width': 'max-content'})
    button = widgets.Button(description='Pārbaudīt', button_style='info')
    feedback = widgets.HTML()

    def on_click(b):
        attempts[index] += 1
        if radio.index == correct_index:
            penalty = 0.05 * (attempts[index] - 1)
            results[index] = max(1 - penalty, 0)
            feedback.value = "<span style='color:green'>✔️ Pareizi!</span>"
            radio.disabled = True
            button.disabled = True
        else:
            feedback.value = "<span style='color:red'>❌ Nepareizi. Mēģini vēlreiz.</span>"

    button.on_click(on_click)
    with out:
        display(widgets.HTML(f"<b>{qdata['question']}</b>"), radio, button, feedback)
    return out

for i, q in enumerate(questions_data):
    widget = create_question_widget(i, q)
    outputs.append(widget)
    display(widget)

submit_btn = widgets.Button(description="✅ Iesniegt testu", button_style='success')
result_html = widgets.HTML()

def on_submit_clicked(b):
    if any(r is None for r in results):
        result_html.value = "<b style='color:red'>⚠️ Lūdzu, atbildi uz visiem jautājumiem pirms iesniegšanas.</b>"
    else:
        score = round(sum(results) / len(results) * 100)
        status = "✅ Nokārtots!" if score >= 70 else "❌ Nenokārtots."
        color = "green" if score >= 70 else "red"
        result_html.value = f"<h3>Rezultāts: <span style='color:{color}'>{score}%</span> – {status}</h3>"

submit_btn.on_click(on_submit_clicked)
display(widgets.VBox([submit_btn, result_html]))
```
