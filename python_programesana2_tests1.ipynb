```python
import ipywidgets as widgets
from IPython.display import display
import random

results = []
attempts = []

questions = [
    ("1. Ko dara pygame.init()?", [
        "InicializÄ“ visus importÄ“tos pygame moduÄ¼us",
        "PalaiÅ¾ spÄ“les ciklu",
        "IelÄdÄ“ spÄ“les logu",
        "Izveido galveno varoÅ†a sprite objektu"
    ], 0),
    ("2. Kura metode tiek izmantota, lai uzzÄ«mÄ“tu taisnstÅ«ri pygame logÄ?", [
        "pygame.draw.rect()",
        "pygame.screen.rectangle()",
        "pygame.create.rect()",
        "pygame.render.rect()"
    ], 0),
    ("3. KÄ tiek atjauninÄts logs pygame pÄ“c zÄ«mÄ“Å¡anas?", [
        "pygame.display.flip()",
        "pygame.update.display()",
        "pygame.screen.refresh()",
        "pygame.render.frame()"
    ], 0),
    ("4. Ko pÄrbauda event.type == pygame.QUIT?", [
        "Vai lietotÄjs ir nospiedis loga aizvÄ“rÅ¡anas pogu",
        "Vai lietotÄjs nospieda taustiÅ†u Enter",
        "Vai spÄ“lei jÄsÄkas no jauna",
        "Vai spÄ“lÄ“tÄjs zaudÄ“ja"
    ], 0),
    ("5. KurÅ¡ modulis tiek izmantots, lai veidotu GUI lietotnes Python valodÄ?", [
        "tkinter",
        "pygame.gui",
        "guimodule",
        "tkbuilder"
    ], 0),
    ("6. KurÅ¡ tkinter logelements tiek izmantots, lai saÅ†emtu tekstu no lietotÄja?", [
        "Entry",
        "Label",
        "Button",
        "Canvas"
    ], 0),
    ("7. Ko dara mainloop() tkinter lietotnÄ“?", [
        "PalaiÅ¾ GUI notikumu ciklu",
        "PieslÄ“dzas datubÄzei",
        "Atjauno Python vidi",
        "Izveido loga izvietojumu"
    ], 0),
    ("8. KurÅ¡ tkinter logelements Ä¼auj zÄ«mÄ“t figÅ«ras kÄ taisnstÅ«rus un tekstu?", [
        "Canvas",
        "Frame",
        "Box",
        "Entry"
    ], 0)
]

random.shuffle(questions)
results = [None] * len(questions)
attempts = [0] * len(questions)

def create_question(index, question, options, correct_index):
    out = widgets.Output()
    option_pairs = list(enumerate(options))
    random.shuffle(option_pairs)
    shuffled_indices, shuffled_options = zip(*option_pairs)
    new_correct_index = shuffled_indices.index(correct_index)

    radio = widgets.RadioButtons(options=shuffled_options, layout={'width': 'max-content'})
    button = widgets.Button(description='PÄrbaudÄ«t')
    feedback = widgets.HTML()

    def on_click(b):
        attempts[index] += 1
        if radio.index == new_correct_index:
            penalty = 0.05 * (attempts[index] - 1)
            results[index] = max(1 - penalty, 0)
            feedback.value = "<span style='color:green'>âœ”ï¸ Pareizi!</span>"
            radio.disabled = True
            button.disabled = True
        else:
            feedback.value = "<span style='color:red'>âŒ Nepareizi. MÄ“Ä£ini vÄ“lreiz.</span>"

    button.on_click(on_click)
    with out:
        display(widgets.HTML(f"<b>{question}</b>"), radio, button, feedback)
    display(out)

for i, (q, opts, correct) in enumerate(questions):
    create_question(i, q, opts, correct)

def calculate_score():
    completed = all(r is not None for r in results)
    if not completed:
        return widgets.HTML("<b style='color:red'>âš ï¸ LÅ«dzu, atbildi uz visiem jautÄjumiem pirms iesniegÅ¡anas.</b>")
    score = round(sum(results) / len(results) * 100)
    status = "âœ… NokÄrtots!" if score >= 70 else "âŒ NenokÄrtots."
    color = "green" if score >= 70 else "red"
    return widgets.HTML(f"<h3>RezultÄts: <span style='color:{color}'>{score}%</span> â€“ {status}</h3>")

display(calculate_score())
```
