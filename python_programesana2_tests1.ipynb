Izvēlies pareizās atbildes. Katru reizi jautājumu un atbilžu secība tiek sajaukta.


```python
import ipywidgets as widgets
from IPython.display import display
import random

results = []
attempts = []

questions = [
    ("1. Ko dara pygame.init()?", [
        "Inicializē visus importētos pygame moduļus",
        "Palaiž spēles ciklu",
        "Ielādē spēles logu",
        "Izveido galveno varoņa sprite objektu"
    ], 0),
    ("2. Kura metode tiek izmantota, lai uzzīmētu taisnstūri pygame logā?", [
        "pygame.draw.rect()",
        "pygame.screen.rectangle()",
        "pygame.create.rect()",
        "pygame.render.rect()"
    ], 0),
    ("3. Kā tiek atjaunināts logs pygame pēc zīmēšanas?", [
        "pygame.display.flip()",
        "pygame.update.display()",
        "pygame.screen.refresh()",
        "pygame.render.frame()"
    ], 0),
    ("4. Ko pārbauda event.type == pygame.QUIT?", [
        "Vai lietotājs ir nospiedis loga aizvēršanas pogu",
        "Vai lietotājs nospieda taustiņu Enter",
        "Vai spēlei jāsākas no jauna",
        "Vai spēlētājs zaudēja"
    ], 0),
    ("5. Kurš modulis tiek izmantots, lai veidotu GUI lietotnes Python valodā?", [
        "tkinter",
        "pygame.gui",
        "guimodule",
        "tkbuilder"
    ], 0),
    ("6. Kurš tkinter logelements tiek izmantots, lai saņemtu tekstu no lietotāja?", [
        "Entry",
        "Label",
        "Button",
        "Canvas"
    ], 0),
    ("7. Ko dara mainloop() tkinter lietotnē?", [
        "Palaiž GUI notikumu ciklu",
        "Pieslēdzas datubāzei",
        "Atjauno Python vidi",
        "Izveido loga izvietojumu"
    ], 0),
    ("8. Kurš tkinter logelements ļauj zīmēt figūras kā taisnstūrus un tekstu?", [
        "Canvas",
        "Frame",
        "Box",
        "Entry"
    ], 0)
]

random.shuffle(questions)
results = [None] * len(questions)
attempts = [0] * len(questions)

def create_question(index, question, options, correct_index):
    out = widgets.Output()
    option_pairs = list(enumerate(options))
    random.shuffle(option_pairs)
    shuffled_indices, shuffled_options = zip(*option_pairs)
    new_correct_index = shuffled_indices.index(correct_index)

    radio = widgets.RadioButtons(options=shuffled_options, layout={'width': 'max-content'})
    button = widgets.Button(description='Pārbaudīt')
    feedback = widgets.HTML()

    def on_click(b):
        attempts[index] += 1
        if radio.index == new_correct_index:
            penalty = 0.05 * (attempts[index] - 1)
            results[index] = max(1 - penalty, 0)
            feedback.value = "<span style='color:green'>✔️ Pareizi!</span>"
            radio.disabled = True
            button.disabled = True
        else:
            feedback.value = "<span style='color:red'>❌ Nepareizi. Mēģini vēlreiz.</span>"

    button.on_click(on_click)
    with out:
        display(widgets.HTML(f"<b>{question}</b>"), radio, button, feedback)
    display(out)

for i, (q, opts, correct) in enumerate(questions):
    create_question(i, q, opts, correct)

def calculate_score():
    completed = all(r is not None for r in results)
    if not completed:
        return widgets.HTML("<b style='color:red'>⚠️ Lūdzu, atbildi uz visiem jautājumiem pirms iesniegšanas.</b>")
    score = round(sum(results) / len(results) * 100)
    status = "✅ Nokārtots!" if score >= 70 else "❌ Nenokārtots."
    color = "green" if score >= 70 else "red"
    return widgets.HTML(f"<h3>Rezultāts: <span style='color:{color}'>{score}%</span> – {status}</h3>")

display(calculate_score())
```
