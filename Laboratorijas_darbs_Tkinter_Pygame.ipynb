### 📚 **Uzdevums**

Tu veidosi vienkāršu `pygame` lietotni, kur **bumbiņa atlēks logā**. Ar `tkinter` palīdzību lietotājs varēs mainīt:

- Bumbiņas **krāsu**
- Bumbiņas **izmēru (rādiusu)**
- Bumbiņas **ātrumu**

Programma:
- Izmantos `tkinter` iestatījumu izvēlei
- Zīmēs bumbiņu ar `pygame`
- Uzsāks spēli ar izvēlētajiem iestatījumiem, kad nospiests **Start Game**


```
import pygame
import tkinter as tk
from tkinter import ttk

# Noklusētās vērtības
ball_color = (255, 0, 0)
ball_radius = 30
ball_speed = 5

def start_game():
    global ball_color, ball_radius, ball_speed

    ball_radius = int(size_var.get())
    ball_speed = int(speed_var.get())

    color = color_var.get()
    if color == "Sarkana":
        ball_color = (255, 0, 0)
    elif color == "Zaļa":
        ball_color = (0, 255, 0)
    elif color == "Zila":
        ball_color = (0, 0, 255)

    pygame.init()
    width, height = 500, 400
    screen = pygame.display.set_mode((width, height))
    pygame.display.set_caption("Atlēcošās bumbiņas spēle")
    clock = pygame.time.Clock()

    x, y = 100, 100
    dx, dy = ball_speed, ball_speed

    running = True
    while running:
        screen.fill((255, 255, 255))
        pygame.draw.circle(screen, ball_color, (x, y), ball_radius)

        x += dx
        y += dy

        if x - ball_radius <= 0 or x + ball_radius >= width:
            dx = -dx
        if y - ball_radius <= 0 or y + ball_radius >= height:
            dy = -dy

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        pygame.display.flip()
        clock.tick(60)

    pygame.quit()

# Tkinter logs
root = tk.Tk()
root.title("Spēles iestatījumi")

tk.Label(root, text="Bumbiņas krāsa:").pack()
color_var = tk.StringVar(value="Sarkana")
ttk.Combobox(root, textvariable=color_var, values=["Sarkana", "Zaļa", "Zila"]).pack()

tk.Label(root, text="Bumbiņas izmērs:").pack()
size_var = tk.StringVar(value="30")
tk.Entry(root, textvariable=size_var).pack()

tk.Label(root, text="Bumbiņas ātrums:").pack()
speed_var = tk.StringVar(value="5")
tk.Entry(root, textvariable=speed_var).pack()

tk.Button(root, text="Start Game", command=start_game).pack(pady=10)

root.mainloop()
```

- Ja ātrums = **10**, bumbiņa kustēsies daudz ātrāk un lēks intensīvāk.
